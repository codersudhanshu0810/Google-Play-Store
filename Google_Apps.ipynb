# Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Data

```python
df = pd.read_csv('google_play_apps.csv')
df
```

# Dealing with null data

```python
df.isnull().sum()
```

```python
df = df.dropna()
df
```

```python
df.info()
```

```python
df['Price'] = df['Price'].str.replace('$', '')
df
```

```python
df.info()
```

# Type Conversion

```python
df['Price'] = df['Price'].astype(float)
df
```

```python
df.info()
```

```python
df['Rating'] = df['Rating'].astype(float)
df
```

```python
df['Reviews'] = df['Reviews'].astype(int)
df
```

```python
df['Installs'] = df['Installs'].str.replace('+', '')
df
```

```python
df.info()
```

```python
df['Installs'] = df['Installs'].str.replace(',', '')
df
```

```python
df['Installs'] = df['Installs'].astype(int)
df
```

# Analyse the Data
## Creating new column 'AmountPaid'

```python
df['AmountPaid'] = df['Price'] * df['Installs']
df
```

# Data Visualization

```python
gdf = df['Category'].value_counts().head(15)
gdf = gdf.reset_index()
gdf
```

```python
plt.figure(figsize = (30, 5))
plt.bar(gdf['Category'], gdf['count'])
plt.xlabel('Category Wise Count')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
pdf = df.groupby('Category').agg(Paid = ('AmountPaid', 'sum')).head(10)
pdf = pdf.reset_index().sort_index(ascending = False)
pdf
```

```python
plt.figure(figsize = (20, 5))
plt.bar(pdf['Category'], pdf['Paid'])
plt.xlabel('Category Wise Paid Apps')
plt.ylabel('Paid Amount')
plt.grid()
plt.show()
```
